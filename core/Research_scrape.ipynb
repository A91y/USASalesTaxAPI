```python
def clean_state_name(stateName: str) -> str:
    formatted = []
    stateName = stateName.split(" ")
    for i in stateName:
        formatted.append(i.capitalize())
    stateName = " ".join(formatted)
    return stateName
```

```python
print(clean_state_name("north carolina"))
```
